##Lakukan deteksi bagian wajah gambar mjordan.jpg, kemudian tindih bagian wajah tersebut dengan gambar smile.png.

```python
jordan2 = cv.imread('/content/drive/MyDrive/images/facedet/mjordan.jpg')
emoji = cv.imread('/content/drive/MyDrive/images/facedet/Salinan smile.png', cv.IMREAD_UNCHANGED)

gray3 = cv.cvtColor(jordan2, cv.COLOR_BGR2GRAY)

jordan2_wajah = cascade_wajah.detectMultiScale(gray3, scaleFactor=1.1, minNeighbors=5)

for (x, y, w, h) in jordan2_wajah:
    emoji_resized = cv.resize(emoji, (w, h))
    overlay = cv.addWeighted(jordan2[y:y+h, x:x+w], 1, emoji_resized[:, :, 0:3], 1, 0)
    jordan2[y:y+h, x:x+w] = overlay

plt.figure(figsize=(12, 6))
plt.imshow(cv.cvtColor(jordan2, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

##Deteksi wajah dengan menggunakan bounding box berwarna hitam, deteksi mata menggunakan bounding box berwarna putih, dan deteksi mata kanan menggunakan bounding box berwarna merah. Gunakan gambar people.jpg untuk mengerjakan soal pada nomor ini.

```python
face_cascade = cv.CascadeClassifier('/content/drive/MyDrive/images/haarcascades/haarcascade_frontalface_alt.xml')
eye_cascade = cv.CascadeClassifier('/content/drive/MyDrive/images/haarcascades/haarcascade_eye.xml')

gambar = cv.imread('/content/drive/MyDrive/images/facedet/Salinan people.jpg')

gambar_gray = cv.cvtColor(gambar, cv.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gambar_gray, scaleFactor=1.1, minNeighbors=5)

for (x, y, w, h) in faces:
    cv.rectangle(gambar, (x, y), (x+w, y+h), (0, 0, 0), 1)
    roi_gray = gambar_gray[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray, scaleFactor=1.1, minNeighbors=5)
    for (ex, ey, ew, eh) in eyes:
        cv.rectangle(gambar, (x+ex, y+ey), (x+ex+ew, y+ey+eh), (0, 0, 255), 1)
        if ex + ew/2 > w/2:
            cv.rectangle(gambar, (x+ex, y+ey), (x+ex+ew, y+ey+eh), (255, 255, 255), 1)

plt.figure(figsize=(12, 6))
plt.imshow(cv.cvtColor(gambar, cv.COLOR_BGR2RGB))
plt.title('Result')
plt.axis('off')
plt.show()
```
